```python
import json
import time
import pandas as pd


try:

    #*#*#*#*#*#*#*#*#*#*#*#            
    ###STARTING POINT         


    inv_opn=open("data.json")
    inv=inv_opn.read()
    inv_opn.close()

    print("")
    #print(inv)
    ########################
    #Json to dict 
    data=json.loads(inv)
    #print(data)
    #print(data["11120"])

    id_=[]
    name_=[]
    brand_=[]
    price_=[]
    weight_=[]
    quantity_=[]

    for lst_pr in data:
        id_.append(lst_pr)
        name_.append(data[lst_pr]["Name"])
        brand_.append(data[lst_pr]["Brand"])
        price_.append(data[lst_pr]["Price"])
        weight_.append(data[lst_pr]["Weight"])
        quantity_.append(data[lst_pr]["Quantity"])

    data_dict={'ID':id_,' Product Name':name_,' Product Brand':brand_,' Product Price':price_,' Product Weight':weight_,' Product Quantity':quantity_}


    df=pd.DataFrame(data_dict)
    df=df.to_string(index=False)
    print(df)
    print("\n")

    ###############################################################
    #Customer data

    while True:
        c_name=input("\nEnter your name Please: ")
        c_name=c_name.strip()
        if c_name== "" :
            print("Please dont leave this field blank.\n")
            continue
        break

    while True:

        p_id=input("\n"+c_name+" Please Enter the Product id: ")

        if p_id not in data.keys():
            print("\n*NOTE "+c_name+" Please Enter the Correct Product_Id for Shopping\n")
            continue

        break



    while True:
        p_qnt=input("\n"+c_name+" Please Enter the Product Quantity: ")

        qnt_1=data[p_id]["Quantity"]
            
        try: p_qnt=int(p_qnt)
                
        except:
            print("\n*NOTE "+c_name+" Please Enter Quantity in Whole NUMBER.\n")
            continue
                
        if 0>p_qnt :
            print("\n*NOTE "+c_name+" Please Enter Quantity in Positive NUMBER\n")
            continue
        if p_qnt == 0 :
            print("\n*NOTE "+c_name+" Please Select atleast One Quantity \n")
            continue

        if p_qnt > qnt_1:

            print("\nWe are so sorry, We don't have That much product Quantity regarding "+p_id+" ID.")
            print("Please Choose Product Quantity from provided Chart.\n")
            print(df)
            print("\nYour Choosen Product Id is: "+p_id)
            continue

        break




    print("\n")


                
    #print(type(data["11120"]["Name"]))
    nm_1=data[p_id]["Name"]
    prc_1=data[p_id]["Price"]
    brnd_1=data[p_id]["Brand"]

    #print(nm_1)
    #print(prc_1)

    #tran_id=""
    ################################################################
    #Bill json creation
    #dict

    bl_tm=time.ctime()

    try:
        sl_opn=open("sales.json")
        sale=sl_opn.read()
        sl_opn.close()
        cstmr_data=json.loads(sale)

        tran_id=str(len(cstmr_data)+1)

        cstmr_data[tran_id]={"Billing Time": bl_tm , "Customer Name" : c_name , "Product Id" : p_id , "Total purchase" : str(p_qnt)+" "+brnd_1+" "+nm_1 , "Price of Each Product" : str(prc_1) , "Bill Amount" : str(p_qnt*prc_1)+" Rs." }

    except:

        tran_id=str(1)

        cstmr_data=dict()
        cstmr_data[tran_id]={"Billing Time": bl_tm , "Customer Name" : c_name , "Product Id" : p_id , "Total purchase" : str(p_qnt)+" "+brnd_1+" "+nm_1 , "Price of Each Product" : str(prc_1) , "Bill Amount" : str(p_qnt*prc_1)+" Rs." }



    ##########################################################################################


    #########################
    #Generate bill
    print("#############################################")
    print("")
    bl_name="\nCustomer Name: "+c_name
    bl_prdct="You have Byed: "+str(p_qnt)+" "+brnd_1+" "+nm_1
    bl_pay="Total Amount to pay: "+"₹"+str(p_qnt*prc_1)
    bl_prc="Price of Each Product: "+str(prc_1)
    bl_pr_id="Product Id : "+str(p_id)


    tran_id="Transaction ID "+tran_id
    print(tran_id)
    print(bl_tm)
    print(bl_pr_id)
    print(bl_name)
    print(bl_prc)
    print(bl_prdct)
    print("--------------------------------")
    print("|                              |")
    print("   ",end="")
    print(bl_pay)
    print("|                              |")
    print("--------------------------------")
    print("")

    print("#############################################")



    #print(type(cstmr_data))

    #json
    cstmr_data=json.dumps(cstmr_data,indent = 1)

    #print(cstmr_data)

    billjs_2_fl=open("sales.json",'w')

    bill_nw_js=billjs_2_fl.write(cstmr_data)

    billjs_2_fl.close()





    ###############

    #update data
    data[p_id]["Quantity"]=data[p_id]["Quantity"]-p_qnt
    #print(data)
    #print("\n")






    #################################
    #converting data --------- dict to json
    data=json.dumps(data,indent = 1)
    #print(type(data))
    #print(data)

    js_2_fl=open("data.json",'w')
    nw_js=js_2_fl.write(data)

    js_2_fl.close()
                
except:
    print("\nWE are so Sorry Our INVENTORY is in under DATA PROCESSING.")
  
```
