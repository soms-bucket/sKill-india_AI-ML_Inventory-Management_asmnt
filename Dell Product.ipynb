```python
import json
import time
import pandas as pd
from string import capwords as capi

try:

	#*#*#*#*#*#*#*#*#*#*#*#            
	###STARTING POINT         


	inv_opn=open("data.json")
	inv=inv_opn.read()
	inv_opn.close()

	print("")
	#print(inv)
	########################
	#Json to dict 
	data=json.loads(inv)
	#print(type(data))
	#print(data["11120"])

	id_=[]
	name_=[]
	brand_=[]
	price_=[]
	weight_=[]
	quantity_=[]

	for lst_pr in data:
	    id_.append(lst_pr)
	    name_.append(data[lst_pr]["Name"])
	    brand_.append(data[lst_pr]["Brand"])
	    price_.append(data[lst_pr]["Price"])
	    weight_.append(data[lst_pr]["Weight"])
	    quantity_.append(data[lst_pr]["Quantity"])

	data_dict={'ID':id_,' Product Name':name_,' Product Brand':brand_,' Product Price':price_,' Product Weight':weight_,' Product Quantity':quantity_}


	df=pd.DataFrame(data_dict)
	df=df.to_string(index=False)
	print(df)
	print("\n")



	def del_data(data):
		for i in range(1):

			c_name="ADMIN"

			##############################################################
			#Product id

			while True:
				p_id=input("\nTo DELETE Please Enter the Product id: ")

				try: p_id=int(p_id)

				except: 
					print("please Enter Product ID in whole NUMBER.")
					continue

				if 0>p_id :
					print("\n*NOTE "+c_name+" Please Enter Quantity in Positive NUMBER\n")
					continue

				p_id=str(p_id)
				
				if p_id in data.keys():

					print(p_id+" Details is HEre\n\n")
					print("Product ID: "+p_id)
					print("Product Name: "+data[p_id]["Name"])
					print("Brand Name: "+data[p_id]["Brand"])
					print("Product Price: "+str(data[p_id]["Price"]))
					print("Product Weight: "+str(data[p_id]["Weight"]))
					print("Product Quantity: "+str(data[p_id]["Quantity"]))
					print("\n")

					#*#*#*#*#*#*#*#*#*#*#*#*#*
					data.pop(p_id)
					print("\n Successfull Deleted "+p_id+" ID")

					print("\nHere is Your New Product Data Chart.\n")
					id_=[]
					name_=[]
					brand_=[]
					price_=[]
					weight_=[]
					quantity_=[]

					for lst_pr in data:
					    id_.append(lst_pr)
					    name_.append(data[lst_pr]["Name"])
					    brand_.append(data[lst_pr]["Brand"])
					    price_.append(data[lst_pr]["Price"])
					    weight_.append(data[lst_pr]["Weight"])
					    quantity_.append(data[lst_pr]["Quantity"])

					data_dict={'ID':id_,' Product Name':name_,' Product Brand':brand_,' Product Price':price_,' Product Weight':weight_,' Product Quantity':quantity_}


					df=pd.DataFrame(data_dict)
					df=df.to_string(index=False)
					print(df)
					print("\n")
					


					#################################
					#converting data --------- dict to json
					data=json.dumps(data,indent = 1)
					#print(type(data))
					#print(data)

					js_2_fl=open("data.json",'w')
					nw_js=js_2_fl.write(data)

					js_2_fl.close()



				
				else:
					print("\nplease Choose correct Product ID")
					continue	


					
					#break

				break



	del_data(data)

except:
	print("\nHere is no DATA to DELETE")
```
